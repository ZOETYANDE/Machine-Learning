```python
import pandas as pd;
import matplotlib.pyplot as plt
import seaborn as sns

path = "./Datasets/bill.csv"

df = pd.read_csv(path);

df.info()
```

## Bar chart

```python
# Bar chart with day against reward
plt.bar(df['day'], df['reward'])

plt.title('Bar Chart using Matplotlib')

# Setting the X and Y labels
plt.xlabel('Day')
plt.ylabel('Reward')

#Adding the legends
plt.show()
```

```python
# Bar Chart width day against reward using Seaborn
sns.barplot(x='day', y='reward', data=df, hue='gender')
plt.title("Bar Chart using Seaborn")
plt.show()
```

## Line Chart

```python
# Line Chart using Matplotlib
plt.plot(df['reward'])
plt.plot(df['person'])

#Adding Title to the plot
plt.title("Line Plot using Matplotlib")

# Setting the X and Y labels
plt.xlabel('reward')
plt.ylabel('Person')

plt.show()
```

```python
# Line Chart using Seaborn
sns.lineplot(data=df.drop(['total_bill'], axis=1))
plt.title("Line Plot Using Seaborn")
plt.show()
```

## Pie Chart

```python
# Creating dataset 
courses = ['Database', 'AI', 'Python', 'Cyber_Security', 'Networks']
data = [18,26,8,11,36]

# Creating plot
fig = plt.figure(figsize=(10, 7))
plt.pie(data, labels=courses)
plt.title("Pie Chart using Matplotlib")
```

```python
# Let's take the same dataset as the one above

# Creating plot
fig = plt.figure(figsize=(10, 7))
plt.title("Pie chart using Seaborn")

# define Seaborn color palette to use
palette_color = sns.color_palette('bright')

# plotting data on chart
plt.pie(data, labels=courses,colors=palette_color, autopct='%.0f%%')

# displaying chart

plt.show()
```

```python
# Declaring exploding pie
expolde = [0, 0.1, 0.1, 0.2, 0.3]

# Define Seaborn color palette to use
palette_color = sns.color_palette('dark')

#ploting data on chart
plt.pie(data, labels=courses, colors=palette_color, explode=expolde, autopct='%.0f%%')


#displaying chart
plt.show()
```

## Histogram

```python
# histogram of total_bills
plt.hist(df['total_bill'])

plt.title("Histogram using Matplotlib")

# Adding the legends
plt.show()
```

```python
sns.histplot(x='total_bill', data=df, kde=True, hue='gender')
plt.title("Histogram using Seaborn")

plt.show()
```

## Scatter plot

```python
# Scatter plot with Day against reward using Matplotlib
plt.scatter(df['day'], df['reward'])

#To add Title to the plot
plt.title("Scatter Plot using Matplotlib")

# To set the X and Y labels
plt.xlabel("Day")
plt.ylabel("Reward")

plt.show()
```
